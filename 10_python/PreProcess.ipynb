# PDFからデータを抜き取る

```python
import re
import os
import pandas as pd

import pdfplumber

def extract_pdf_content(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        # ページごとにテキストを抽出
        text_per_page = [page.extract_text() for page in pdf.pages]

        # 企業名、年収、業務内容、処遇を抽出
        companys = ''
        incomes = ''
        descriptions = ''
        treatments = ''

        for text in text_per_page:
            # 企業名の抽出
            company = extract_company_name(text)
            companys = company if company is not None else companys
            # companys.append(company)

            # 年収の抽出
            income = extract_annual_income(text)
            incomes = income if income is not None else incomes
            # incomes.append(income)

            # 業務内容の抽出
            description = extract_job_description(text)
            descriptions = description if description is not None else descriptions
            # descriptions.append(description)

            # 処遇の抽出
            treatment = extract_treatment(text)
            treatments = treatment if treatment is not None else treatments
            # treatments.append(treatment)

        # データフレームに格納
        # df = pd.DataFrame({
        #     'company': companys,
        #     'income': incomes,
        #     'description': descriptions,
        #     'treatment': treatments
        # },index = [0])
        info = [companys,incomes,descriptions,treatments]

        # return text_per_page
        return info

def extract_company_name(text):
    match = re.search(r'企業名\s+(.+?)\s', text)
    return match.group() if match else None

def extract_annual_income(text):
    match = re.search(r'入社時想定年収 (.+?)選考プロセス', text, re.DOTALL)
    return match.group() if match else None

def extract_job_description(text):
    match = re.search(r'(仕事内容|業務内容)(.+?)(必須要件|会社・仕事の魅力)', text, re.DOTALL)
    return match.group(1).strip() if match else None

def extract_treatment(text):
    match = re.search(r'勤務地住所 (.+?)休日休暇', text, re.DOTALL)
    return match.group().strip() if match else None


def create_summary(folder_path):
    summary = []
    for filename in os.listdir(folder_path):
        if filename.endswith(".pdf"):
            pdf_path = os.path.join(folder_path, filename)
            text_per_page = extract_pdf_content(pdf_path)

            summary.append({'filename': filename, 'company_name': text_per_page[0], 'income': text_per_page[1], 'description': text_per_page[2], 'treatment': text_per_page[3]})
    return summary

# フォルダのパスを指定
folder_path = r'C:\Users\nishi\OneDrive\ドキュメント\10.LIFE\20_実績\2023_就活\ギークニア\転職票'
outpuf_file = r'PDFlist.csv'
output_path = os.path.join(folder_path, outpuf_file)
summary = create_summary(folder_path)

# データフレームに変換
df = pd.DataFrame(summary)

df.to_csv(output_path,encoding='cp932', index=False)

# データフレームの表示
print(df)
```
