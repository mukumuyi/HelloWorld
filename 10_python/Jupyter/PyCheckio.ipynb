```python
# GitHub/Merge Intervals
def merge_intervals(intervals):
    # Your code here
    list_intervals = []
    if len(intervals) == 0: return list_intervals
    tmp = list(intervals[0])

    for i in intervals:
        if i[0] >= tmp[0] and i[0] <= tmp[1] + 1:
            if i[1] > tmp[1]:
                tmp[1] = i[1]
        else:
            list_intervals.append(tuple(tmp))
            tmp = list(i)

    list_intervals.append(tuple(tmp))
    print(list_intervals)

    return list_intervals

if __name__ == '__main__':
    #These "asserts" using only for self-checking and not necessary for auto-testing
    assert merge_intervals([(1, 4), (2, 6), (8, 10), (12, 19)]) == [(1, 6), (8, 10), (12, 19)], "First"
    assert merge_intervals([(1, 12), (2, 3), (4, 7)]) == [(1, 12)], "Second"
    assert merge_intervals([(1, 5), (6, 10), (10, 15), (17, 20)]) == [(1, 15), (17, 20)], "Third"
    print('Done! Go ahead and Check IT')
```

```python
# O'Relly/The Flat Dictionary
# Recurse Logic
# Dict Loop

def flatten(dictionary):
    # your code here
    keys = dictionary.keys()
    values = dictionary.values()

    result_dict = {}

    for k in keys:
        v = dictionary.get(k)
        if len(v) == 0:
            result_dict[k] = ""
        elif isinstance(v, dict):
            for k1,v1 in flatten(v).items():
                result_dict[k + "/" + k1] = v1
        else :
            result_dict[k] = v

    return result_dict


if __name__ == '__main__':
    test_input ={"empty": {}}
    #test_input = {"key": {"deeper": {"more": {"enough": "value"}}}}
    print(' Input: {}'.format(test_input))
    print('Output: {}'.format(flatten(test_input)))

    #These "asserts" using only for self-checking and not necessary for auto-testing
    #ssert flatten({"key": "value"}) == {"key": "value"}, "Simple"
    assert flatten(
        {"key": {"deeper": {"more": {"enough": "value"}}}}
    ) == {"key/deeper/more/enough": "value"}, "Nested"
    assert flatten({"empty": {}}) == {"empty": ""}, "Empty value"
    assert flatten({"name": {
                        "first": "One",
                        "last": "Drone"},
                    "job": "scout",
                    "recent": {},
                    "additional": {
                        "place": {
                            "zone": "1",
                            "cell": "2"}}}
    ) == {"name/first": "One",
          "name/last": "Drone",
          "job": "scout",
          "recent": "",
          "additional/place/zone": "1",
          "additional/place/cell": "2"}
    print('You all set. Click "Check" now!')
```
